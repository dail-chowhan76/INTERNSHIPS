```python
"""
MNIST Digit Recognition System
Name : MUHAMMAD DAIL CHOWHAN
Email : dailchowhan@gmail.com
Date: 2 January'2026
Intern ID : ARCH-2610-0765
"""

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.utils import to_categorical


plt.style.use('default')
sns.set_palette("husl")

print("=" * 70)
print("MNIST DIGIT RECOGNITION SYSTEM")
print("ARCH TECHNOLOGIES - Task 2 Submission")
print("=" * 70)

print("\nSTEP 1: Loading MNIST Dataset...")
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f"   Training data shape: {X_train.shape}")
print(f"   Testing data shape: {X_test.shape}")
print(f"   Number of classes: {len(np.unique(y_train))}")
print(f"   Image dimensions: {X_train.shape[1]}x{X_train.shape[2]} pixels")

print("\nSTEP 2: Preprocessing Images...")
X_train = X_train.astype('float32') / 255.0
X_test = X_test.astype('float32') / 255.0
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

print("   Images normalized (0 to 1)")
print("   Labels one-hot encoded")

print("\nSTEP 3: Visualizing Sample Data...")

fig, axes = plt.subplots(2, 5, figsize=(12, 5))
axes = axes.ravel()

for i in range(10):
    axes[i].imshow(X_train[i], cmap='gray')
    axes[i].set_title(f"Label: {y_train[i]}")
    axes[i].axis('off')

plt.suptitle('Sample MNIST Handwritten Digits', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('mnist_sample_digits.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nClass Distribution:")
unique, counts = np.unique(y_train, return_counts=True)
for digit, count in zip(unique, counts):
    percentage = (count/len(y_train))*100
    print(f"   Digit {digit}: {count} samples ({percentage:.1f}%)")

print("\nSTEP 4: Building Neural Network Model...")

model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(256, activation='relu'),
    Dropout(0.3),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

print("   Model Architecture:")
model.summary()

print("\nSTEP 5: Training Model...")
print("   (This may take a few minutes)")

history = model.fit(
    X_train, y_train_cat,
    epochs=10,
    batch_size=64,
    validation_split=0.2,
    verbose=1
)

print("   Model training completed!")

print("\nSTEP 6: Visualizing Training Progress...")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

ax1.plot(history.history['accuracy'], label='Training Accuracy', linewidth=2)
ax1.plot(history.history['val_accuracy'], label='Validation Accuracy', linewidth=2)
ax1.set_title('Model Accuracy Over Epochs', fontsize=12, fontweight='bold')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2.plot(history.history['loss'], label='Training Loss', linewidth=2)
ax2.plot(history.history['val_loss'], label='Validation Loss', linewidth=2)
ax2.set_title('Model Loss Over Epochs', fontsize=12, fontweight='bold')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Loss')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.suptitle('Training History', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('training_history.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nSTEP 7: Evaluating Model Performance...")

test_loss, test_accuracy = model.evaluate(X_test, y_test_cat, verbose=0)
print(f"   Test Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"   Test Loss: {test_loss:.4f}")

y_pred_probs = model.predict(X_test, verbose=0)
y_pred = np.argmax(y_pred_probs, axis=1)

print("\nSTEP 8: Generating Confusion Matrix...")

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='YlOrRd', 
            xticklabels=range(10), yticklabels=range(10),
            cbar_kws={'label': 'Number of Samples'})
plt.title('Confusion Matrix - Digit Recognition', fontsize=14, fontweight='bold')
plt.ylabel('True Digit', fontsize=12)
plt.xlabel('Predicted Digit', fontsize=12)
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nSTEP 9: Detailed Classification Report...")
print("\n" + "="*60)
print("CLASSIFICATION REPORT")
print("="*60)
print(classification_report(y_test, y_pred, digits=4))
print("="*60)

print("\nSTEP 10: Analyzing Misclassifications...")

misclassified_idx = np.where(y_pred != y_test)[0]
correct_idx = np.where(y_pred == y_test)[0]

correct_percentage = (len(correct_idx)/len(X_test))*100
misclassified_percentage = (len(misclassified_idx)/len(X_test))*100

print(f"   Correctly classified: {len(correct_idx)}/{len(X_test)} ({correct_percentage:.2f}%)")
print(f"   Misclassified: {len(misclassified_idx)}/{len(X_test)} ({misclassified_percentage:.2f}%)")

if len(misclassified_idx) > 0:
    print("\n   Top Misclassified Examples:")
    fig, axes = plt.subplots(2, 5, figsize=(14, 6))
    axes = axes.ravel()
    for i in range(min(10, len(misclassified_idx))):
        idx = misclassified_idx[i]
        axes[i].imshow(X_test[idx], cmap='gray')
        axes[i].set_title(f"True: {y_test[idx]} | Pred: {y_pred[idx]}", fontsize=10)
        axes[i].axis('off')
    
    plt.suptitle('Misclassified Digit Examples', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.savefig('misclassified_examples.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("\n   Most Common Confusions:")
    for i in range(10):
        for j in range(10):
            if i != j and cm[i, j] > 5:
                print(f"     Digit {i} -> Digit {j}: {cm[i, j]} times")

print("\nSTEP 11: Testing with Random Samples...")

fig, axes = plt.subplots(2, 5, figsize=(14, 6))
axes = axes.ravel()

for i in range(10):
    random_idx = np.random.randint(0, len(X_test))
    test_image = X_test[random_idx].reshape(1, 28, 28)
    prediction = model.predict(test_image, verbose=0)
    predicted_digit = np.argmax(prediction)
    confidence = np.max(prediction) * 100
    axes[i].imshow(X_test[random_idx], cmap='gray')
    color = 'green' if predicted_digit == y_test[random_idx] else 'red'
    axes[i].set_title(f"True: {y_test[random_idx]}\nPred: {predicted_digit}\nConf: {confidence:.1f}%", 
                     color=color, fontsize=10)
    axes[i].axis('off')

plt.suptitle('Random Test Predictions (Green=Correct, Red=Wrong)', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('random_predictions.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "="*70)
print("PERFORMANCE SUMMARY")
print("="*70)

class_accuracy = []
for digit in range(10):
    mask = y_test == digit
    if np.sum(mask) > 0:
        acc = np.sum(y_pred[mask] == digit) / np.sum(mask)
        class_accuracy.append(acc)
        print(f"Digit {digit}: {acc:.4f} ({acc*100:.2f}%)")

print("-" * 70)
print(f"Overall Accuracy: {test_accuracy:.4f} ({test_accuracy*100:.2f}%)")
print(f"Best Class: Digit {np.argmax(class_accuracy)} ({max(class_accuracy)*100:.2f}%)")
print(f"Worst Class: Digit {np.argmin(class_accuracy)} ({min(class_accuracy)*100:.2f}%)")

print("\n" + "="*70)
print("TASK 2 COMPLETED SUCCESSFULLY!")
print("="*70)
print("\nGenerated Files:")
print("   1. mnist_sample_digits.png    - Sample images from dataset")
print("   2. training_history.png       - Accuracy/Loss plots")
print("   3. confusion_matrix.png       - Confusion matrix visualization")
print("   4. misclassified_examples.png - Error analysis")
print("   5. random_predictions.png     - Random test predictions")
print("\nModel Performance:")
print(f"   Final Test Accuracy: {test_accuracy*100:.2f}%")
print(f"   Misclassification Rate: {misclassified_percentage:.2f}%")
print(f"   Model trained on: {len(X_train):,} images")
print(f"   Model tested on: {len(X_test):,} images")
print("\n" + "="*70)
```
